```python
import pandas as pd
#pandas是處理csv檔案的套件包
#注意:匯入檔案為csv檔的原因是因為乾淨，無須擔心格式問題

Feb = pd.read_csv("data1.csv")
Nov = pd.read_csv("data2.csv")
```

```python
Feb
```

```python
Nov
```

```python
#列出子集合
set_Feb = set(Feb["Name"])
set_Nov = set(Nov["Name"])
```

```python
set_Feb
```

```python
set_Nov
```

```python
#聯集:union
#全部的學生名稱
union = set_Feb | set_Nov
union
```

```python
#交集:intersection
#從11月份持續上課到2月份的學生
intersection = set_Nov & set_Feb
intersection 
```

```python
#差集:subtract1
#11月有上課，但沒有持續上課到2月份的學生
subtract1 = set_Nov - intersection
subtract1
```

```python
#差集:subtract2
#11月沒有上課但在2月份新加入的學生
subtract2 = set_Feb - intersection
subtract2
```

```python
#對稱差集symmetric_difference
#新加入的學生或退出的學生
symmetric_difference = union - intersection 
symmetric_difference
```
